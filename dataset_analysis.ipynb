# Dataset Info

ISOT Fake News Dataset
The ISOT Fake News dataset is a compilation of several thousands fake news and truthful articles, obtained from different legitimate news sites and sites flagged as unreliable by Politifact.com.
The dataset for this analysis can be found <a href="https://www.uvic.ca/engineering/ece/isot/datasets/fake-news/index.php">here </a>

The dataset contains two types of articles fake and real News. This dataset was collected from realworld sources; the truthful articles were obtained by crawling articles from Reuters.com (News
website). As for the fake news articles, they were collected from different sources. The fake news
articles were collected from unreliable websites that were flagged by Politifact (a fact-checking
organization in the USA) and Wikipedia. The dataset contains different types of articles on different
topics, however, the majority of articles focus on political and World news topics. 
The dataset consists of two CSV files. The first file named “True.csv” contains more than 12,600
articles from reuter.com. The second file named “Fake.csv” contains more than 12,600 articles from
different fake news outlet resources. Each article contains the following information: article title, text,
type and the date the article was published on. 
We focused mostly on collecting articles from 2016 to 2017. The data collected were cleaned and
processed, however, the punctuations and mistakes that existed in the fake news were kept in the text.
The following table gives a breakdown of the categories and number of articles per category.

Thre are 21417 articles in the Real_news dataset: 10145 articles are from World_News category

## Real-news dataset

```python
#load True.csv and turn it into a dataframe.
df_true = pd.read_csv('True.csv')
print(len(df_true))
df_true.head()
```

```python
df_true['label'] = 'true'
```

```python
df_true.head()
```

```python
df_true = df_true.drop(['date'], axis = 1)
```

```python
df_true.subject.value_counts()
```

```python
#Find duplicates
duplicateRowsData = df_true[df_true.duplicated()]
duplicateRowsData
```

```python
# Get indexs of the duplicated rows.
duplicateIndex = []
for row in duplicateRowsData.index: 
    duplicateIndex.append(row)
print(duplicateIndex)  
len(duplicateIndex)
```

```python
df_true = df_true.drop(duplicateIndex)
```

```python
df_true.head(2)
```

```python
#Check again for duplicates
duplicateRowsData = df_true[df_true.duplicated()]
duplicateRowsData
```

```python
#set up stopwords, 
STOPWORDS = stopwords.words('english')
STOPWORDS += list(string.punctuation)
STOPWORDS += ['washington', 'reuters']
```

```python
STOPWORDS = set(STOPWORDS)
```

```python
#create a new column 'Clean text' and lower cases.
df_true['Clean_text'] = df_true['text'].apply(lambda txt: txt.lower())
df_true.Clean_text[0]
```

```python
#remove special characters
df_true['Clean_text'] = df_true['Clean_text'].str.replace("[^a-zA-Z' ]","")
#df_true.Clean_text[0]
```

```python
#remove special characters
df_true['Clean_text'] = df_true['Clean_text'].str.replace(re.compile(r"(^| ).( |$)"), " ")
#df_true.Clean_text[0]
```

```python
#tokenize
df_true['Clean_text'] = df_true['Clean_text'].str.split()
#dataset.Clean_text[0]
```

```python
#remove stopwords
df_true['Clean_text'] = df_true['Clean_text'].apply(lambda text: [word for word in text if word not in STOPWORDS])
#df_true.Clean_text[0]
```

```python
df_true.Clean_text[0]
```

```python
#function to replace the n't with not.
def expand_text(text):
    expanded_text = []
    for word in text:
        if re.search("n't", word):
            expanded_text.append(word.split("n't")[0])
            expanded_text.append("not")
        else:
            expanded_text.append(word)
    return expanded_text
```

```python
#replace "n't" with "not"
df_true['Clean_text'] = df_true['Clean_text'].apply(lambda text: expand_text(text))
#df_true.Clean_text[0]
```

```python
wordNetLemmatizer = WordNetLemmatizer()
porterStemmer = PorterStemmer()
```

```python
#Lemanizing words
df_true['Clean_text'] = df_true['Clean_text'].apply(lambda text: [wordNetLemmatizer.lemmatize(word) for word in text])
#df_true.Clean_text[0]
```

```python
#Stemming the words
df_true['Clean_text'] = df_true['Clean_text'].apply(lambda text: [porterStemmer.stem(word) for word in text])
#df_true.Clean_text[0]
```

```python
#Combine words back to texts
df_true['Clean_text'] = df_true['Clean_text'].apply(lambda text: ' '.join(text))
df_true.Clean_text[0]
```

```python

#print(df_true.text[0])
print(df_true.Clean_text[0])
```

```python
#Delete column Text, we don't need it anymore.
cleaned_df_true = df_true.drop('text', axis=1 )
```

```python
cleaned_df_true.head()
```

```python
#Create a list of all the texts.
true_text = []
for text in df_true['Clean_text']:
    true_text.append(text)
```

```python
#Create bag of words for text. Instead of a list of texts now we'll have a list of words.
true_text_bag = ''.join([str(text) for text in df_true['Clean_text']])
#print(positive_tweets_bag)
```

```python
wordcloud = WordCloud(stopwords = STOPWORDS, background_color = "white", max_words = 1000).generate(true_text_bag)
plt.figure(figsize = (10, 6))
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Most frequent words in true news")
```

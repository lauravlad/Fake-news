```python
import json
import requests
from os import makedirs
from os.path import join, exists
from datetime import date, timedelta
```

```python
ARTICLES_DIR = join('tempdata', 'articles')
makedirs(ARTICLES_DIR, exist_ok=True)
```

```python
# Sample URL
#
# http://content.guardianapis.com/search?from-date=2019-01-01&
# to-date=2019-01-31&order-by=newest&show-fields=all&page-size=200
# &api-key=your-api-key-goes-here

#https://content.guardianapis.com/search?q=debate&tag=politics/politics&from-date=2014-01-01&api-key=test

MY_API_KEY = open("/Users/lauravlad/Desktop/keys/the_guardian.txt").read().strip()
API_ENDPOINT = 'https://content.guardianapis.com/search?q=debate&tag=politics/politics'
my_params = {
    'from-date': "",
    'to-date': "",
    'order-by': "newest",
    'show-fields': 'all',
    'page-size': 200,
    'api-key': MY_API_KEY  
}
```

```python
# day iteration from here:
# http://stackoverflow.com/questions/7274267/print-all-day-dates-between-two-dates
start_date = date(2019, 1, 1)
end_date = date(2019, 12, 31)
dayrange = range((end_date - start_date).days + 1)
for daycount in dayrange:
    dt = start_date + timedelta(days=daycount)
    datestr = dt.strftime('%Y-%m-%d')
    fname = join(ARTICLES_DIR, datestr + '.json')
    if not exists(fname):
        # then let's download it
        print("Downloading", datestr)
        all_results = []
        my_params['from-date'] = datestr
        my_params['to-date'] = datestr
        current_page = 1
        total_pages = 1
        while current_page <= total_pages:
            print("...page", current_page)
            my_params['page'] = current_page
            resp = requests.get(API_ENDPOINT, my_params)
            data = resp.json()
            all_results.extend(data['response']['results'])
            # if there is more than one page
            current_page += 1
            total_pages = data['response']['pages']

        with open(fname, 'w') as f:
            print("Writing to", fname)

            # re-serialize it for pretty indentation
            f.write(json.dumps(all_results, indent=2))
```

```python
# Combining all json files to a one data frame
count = 0
for filename in os.listdir("/Users/lauravlad/Class/Fake-news/tempdata/articles/"):
    count+=1
    if count>2:
        file_path = "/Users/lauravlad/Class/Fake-news/tempdata/articles/" + filename
        df_ = pd.read_json(file_path , encoding='utf-8')
        df = pd.concat(objs= [df,df_], axis=0,ignore_index=True)
        
        # To see the progress, print the number of the file
        if count%100 ==0:
            print ("This is the ",count,"th file -->")
```

```python
# Getting bodytext from the dataframe
lst = []
for i in range(df.shape[0]):
    lst.append(df.fields[i]["bodyText"])
```

```python
# Creating a column with bodytext
df["bodyText"] = lst
```

```python
# Getting headlines from the dataframe
lst_head = []
for i in range(df.shape[0]):
        lst_head.append(df.fields[i]["headline"])
```

```python
# Creating a column with headline
df["headline"] = lst_head
```

```python
# Getting rid of the embty bodies
df = df[df.bodyText != ""]
```

```python
# Filtering articles based on topics
df = df[(df.sectionName == 'US news') | (df.sectionName == 'Politics') | (df.sectionName == 'World news')]
```
